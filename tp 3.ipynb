```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures 
from sklearn.metrics import mean_squared_error

X=eph['NIVEL_ED']
y=eph['CAT_OCUP']
X = np.array(X).reshape((-1, 1))
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 101)


def evalua_modelo(modelo,x_train,x_test,y_train,y_test):
    '''
    El modelo transforma la variable independiente y regresa y contra x , x^2
    Input: modelo, muestras de entrenamiento y prueba
    Output: métricas 
    '''
    #Transformaciones
poly = PolynomialFeatures(degree=2, include_bias=False) 

x_train_poly = poly.fit_transform(x_train)
x_test_poly = poly.fit_transform(x_test)  
        
        #Modelo y predicción

modelo=LinearRegression().fit(x_train_poly,x_train, y_train)
    
y_predpoly = modelo.predict(x_test_poly) 
    #Metricas
    
ecm = mean_squared_error(y_predpoly, y_test)
print(ecm)    ##HASTA ACA VA PERFECTO, PERO SALTA ERROR DESPUES "Classification metrics can't handle a mix of multiclass and continuous targets"

matriz_confusion = confusion_matrix(y_test, y_predpoly)
print('Confusion Matrix :')
print(matriz_confusion)
    
accuracy = accuracy_score(y_test, y_predpoly)
print("Precisión: %.3f" % accuracy) 
print()
    
auc = roc_auc_score(y_test, y_predpoly)
print('AUC: %.2f' % auc)
fpr, tpr, thresholds = roc_curve(y_test, y_predpoly)

 display = RocCurveDisplay(fpr=fpr, tpr=tpr, roc_auc=auc, estimator_name='Reg_log')
    display.plot()  
    plt.plot([0, 1], [0, 1], color='red', linestyle='--')
    plt.show() 


   
   
```
